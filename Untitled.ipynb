## Using with jupyer notebook

This is example of using fuzzy plot with jupyer notebook.
We can config to work with jupyernotebook or generate html chart or on plotly server.

```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go

import fuzzy.fuzzy_main as fuz

init_notebook_mode(connected=True)

my_plt = fuz.FuzzyPlotly(fuz.x_values, fuz.y_values, fuz.std)
my_plt.plot()
```
